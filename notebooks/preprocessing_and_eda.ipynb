```python
# Cell 1: Setup
import sys
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Them thu muc src vao duong dan he thong de import duoc
sys.path.append(os.path.abspath(os.path.join('..', 'src')))

from classification_library import AirQualityLoader

# Cau hinh hien thi
%matplotlib inline
plt.rcParams['figure.figsize'] = (15, 5)

print("Setup Complete!")
```

```python
# Cell 2: Load Data
# Duong dan den file zip
ZIP_PATH = os.path.join('..', 'data', 'raw', 'PRSA2017_Data_20130301-20170228.zip')

# Khoi tao loader
loader = AirQualityLoader(ZIP_PATH)

# Load va xu ly
df = loader.preprocess()

# Xem 5 dong dau tien
df.head()
```

```python
# Cell 3: Ve bieu do PM2.5 (EDA)
# Chon 1 tram de ve thu, vi du: Aotizhongxin
station_name = "Aotizhongxin"
df_station = df[df['station'] == station_name]

plt.figure(figsize=(15, 6))
plt.plot(df_station['datetime'], df_station['PM2.5'], label='PM2.5', color='orange', alpha=0.7)
plt.title(f"Dien bien PM2.5 tai tram {station_name} (2013-2017)")
plt.xlabel("Thoi gian")
plt.ylabel("Nong do PM2.5")
plt.legend()
plt.show()
```
